# User se marks lo aur grade do:

# 90+ → A
# 80+ → B
# 70+ → C
# 60+ → D
# warna → Fail

```python
marks = int(input("Enter your marks: "))

if marks >= 90:
    print("Grade: A")
elif marks >= 80:
    print("Grade: B")
elif marks >= 70:
    print("Grade: C")
elif marks >= 60:
    print("Grade: D")
else:
    print("Grade: Fail")
```

# User se day number (1–7) lo:

# 1 → Monday
# 2 → Tuesday
## ...
# 7 → Sunday

```python
day = int(input("Enter day number (1-7): "))

if day == 1:
    print("Monday")
elif day == 2:
    print("Tuesday")
elif day == 3:
    print("Wednesday")
elif day == 4:
    print("Thursday")
elif day == 5:
    print("Friday")
elif day == 6:
    print("Saturday")
elif day == 7:
    print("Sunday")
else:
    print("Invalid input")
```

# User se month number lo:

# 12,1,2 → Winter
# 3,4,5 → Spring
# 6,7,8 → Summer
# warna → Autumn

```python
month = int(input("Enter month number (1-12): "))

if month in [12, 1, 2]:
    print("Winter")
elif month in [3, 4, 5]:
    print("Spring")
elif month in [6, 7, 8]:
    print("Summer")
elif month in [9, 10, 11]:
    print("Autumn")
else:
    print("Invalid input")
```

# Simple calculator:
# User se operator lo (+, -, *, /):

# → addition
# → subtraction
# → multiplication
# / → division
# warna → "Invalid operator"

```python
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
operator = input("Enter operator (+, -, *, /): ")

if operator == "+":
    print("Result:", num1 + num2)
elif operator == "-":
    print("Result:", num1 - num2)
elif operator == "*":
    print("Result:", num1 * num2)
elif operator == "/":
    if num2 != 0:
        print("Result:", num1 / num2)
    else:
        print("Cannot divide by zero")
else:
    print("Invalid operator")
```
